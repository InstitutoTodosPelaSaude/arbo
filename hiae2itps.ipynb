```python
import pandas as pd
from ftfy import fix_text
```

```python
df = pd.read_excel('data/einstein/Dados Dengue - ITPS.xlsx')
print(df.head())
```

```python
headers = df.columns[0].split(',')
headers
```

```python
def devide_values(row):
    values = row[0].split(',')

    for i in range(len(values)-1):
        if values[i].startswith('"') and not values[i].endswith('"') and values[i+1].endswith('"'):
            values[i] = values[i] + ',' + values[i+1]
            values[i] = values[i].replace('"', '')
            values.pop(i+1)

    assert len(values) == len(headers), f'{len(values)} != {len(headers)}: {values}'


    return [fix_text(value) for value in values]

df_tratado = df.apply(devide_values, axis=1, result_type='expand')
```

```python
df_tratado.columns = headers
df_tratado.sample(10)
```

```python
df_tratado['dh_coleta'] = df_tratado['dh_coleta'].replace('T', ' ', regex=True)
```

```python
df_tratado.to_excel('data/einstein/Dados Dengue - ITPS_tratado.xlsx', index=False)
```
