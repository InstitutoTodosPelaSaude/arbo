### Heatmap

#### Positivos

```python
heat = pd.read_csv("matrix_agegroups_weeks_FLUA_posrate.tsv", sep='\t')
heat.head(3)
```

```python
## Remover colunas com valores NaN e a coluna "country"
heat_cleaned = heat.drop('country', axis=1) #.dropna(axis=1)

## Renomear a coluna "age_group" para "faixas_etarias"
heat_cleaned = heat_cleaned.rename(columns={'age_group': 'faixas etárias'})

## Remover linhas com "Não Detectado" ou "Not tested"
heat_cleaned = heat_cleaned.loc[~heat_cleaned['FLUA_test_result'].str.contains('Não Detectado|Not tested')]

## Substituir as faixas etárias no DataFrame
heat_cleaned['faixas etárias'] = heat_cleaned['faixas etárias'].replace({'0-4': '00-04', '4-9': '04-09', '9-19': '09-19'})

heat_cleaned.head(3)
```

```python
## Derreter (melt) as colunas de data em uma coluna
heat_melted = pd.melt(heat_cleaned, id_vars=['FLUA_test_result', 'faixas etárias'], var_name='semana epidemiológica', value_name='percentual')
heat_melted.head(3)
```

```python
## Transformar os valores da coluna "n" em percentuais
heat_melted['percentual'] = heat_melted['percentual'].apply(lambda x: f'{round(x * 100, 2)}%')
#heat_melted = heat_melted.sort_values('faixas etárias')
heat_melted.head(3)
```

```python
## Filtrar os valores positivos e negativos
heat_melted_positivos = heat_melted[heat_melted['FLUA_test_result'] == 'Pos'].reset_index(drop=True)
# heat_melted_negativos = heat_melted[heat_melted['SC2_test_result'] == 'Negative'].reset_index(drop=True)

## transformar faixas etárias em string
heat_melted_positivos['faixas etárias'] = heat_melted_positivos['faixas etárias'].astype(str)

## transformar semana epidemiológica em dtype striing para facilitar ordenamento
# heat_melted_positivos['semana epidemiológica'] = heat_melted_positivos['semana epidemiológica'].astype(str)

heat_melted_positivos['semana epidemiológica'] = pd.to_datetime(heat_melted_positivos['semana epidemiológica'], format='%Y-%m-%d')
# Remover a parte da hora, mantendo apenas a data
heat_melted_positivos['semana epidemiológica'] = heat_melted_positivos['semana epidemiológica'].dt.date


## verificar dtypes
heat_melted_positivos.dtypes

heat_melted_positivos.head(3)
```

```python
## Escrever o DataFrame final positivo em um arquivo TSV
#heat_melted_positivos.to_csv('figures/_flourish/heatmap_SC2demog.tsv', sep='\t', index=False)

## Escrever o DataFrame final positivo em um arquivo excel -
heat_melted_positivos.to_excel('heatmap_FLUA_demog.xlsx', index=False)
```

#### Estados

```python
## Carregar tabela
heat_ufs = pd.read_csv("heatmap/combined_matrix_state_posrate_full_weeks.tsv", sep='\t')
heat_ufs.head(3)
```

```python
## Correções pontuais
# Converter "RIO DE JANEIRO" para "Rio de Janeiro" em letras minúsculas
heat_ufs['state'] = heat_ufs['state'].replace({'rio de janeiro': 'Rio de Janeiro'})
heat_ufs['state_code'] = heat_ufs['state_code'].replace({'0': 'RJ'})
```

```python
melted_ufs = heat_ufs.melt(id_vars=['state', 'pathogen', 'test_result', 'state_code', 'country'],
                    var_name='semana epidemiológica',
                    value_name='percentual')
melted_ufs.head()
```

```python
## Transformar os valores da coluna "n" em percentuais
melted_ufs['percentual'] = melted_ufs['percentual'].apply(lambda x: f'{round(x * 100, 2)}')
#heat_melted = heat_melted.sort_values('faixas etárias')
melted_ufs.head(3)
```

```python
## Filtrar as linhas
heatmap_sc2 = melted_ufs[melted_ufs['pathogen'] == 'SC2']
```

```python
## Remover e renomear colunas
columns_to_remove = ['state', 'test_result', 'country', 'pathogen']
heatmap_uf_sc2 = heatmap_sc2.drop(columns=columns_to_remove)

# Renomear coluna
heatmap_uf_sc2 = heatmap_uf_sc2.rename(columns={'state_code': 'UF'})
heatmap_uf_sc2.head(10)
```

```python
print(heatmap_uf_sc2.loc[heatmap_uf_sc2['UF'] == 'RJ'])
```

```python
## convertendo 'nan' para  NaN
heatmap_uf_sc2.replace('nan', np.nan, inplace=True)

## Filtrar estados com mais de 80% dos dados preenchidos diferentes de "nan" na coluna "percentual" por estado
porcentagem = heatmap_uf_sc2.groupby('UF')['percentual'].apply(lambda x: (x.notna().sum() / len(x)) * 100)

## Selecionar os estados com mais de 80% das linhas com dados diferentes de "nan"
estados_selecionados = porcentagem[porcentagem > 80].index

## Criar o novo DataFrame contendo apenas os estados selecionados
heatmap_ufs = heatmap_uf_sc2[heatmap_uf_sc2['UF'].isin(estados_selecionados)].dropna()

print(heatmap_ufs)
```

```python
print(heatmap_ufs.loc[heatmap_ufs['UF'] == 'RJ'])
```

```python
## ordenar estados - paliativo enquanto não ajusta para Snakemake
regioes = {
    'Norte': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS']
}

## Mapear as informações de região no DataFrame
heatmap_ufs['Regiao'] = heatmap_ufs['UF'].map({estado: regiao for regiao, estados in regioes.items() for estado in estados})

## Ordenar o DataFrame por região e em ordem alfabética
heatmap_ufs = heatmap_ufs.sort_values(['Regiao', 'UF'])

print(heatmap_ufs)
```

```python
## salvar excel
heatmap_ufs.to_excel('heatmap_states.xlsx', index=False)
```
