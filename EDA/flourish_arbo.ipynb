```python
# heat = pd.read_csv("matrix_agegroups_weeks_FLUA_posrate.tsv", sep='\t')
heat = pd.read_csv("snakemake/results/demography/matrix_agegroups_weeks_DENV_posrate.tsv", sep='\t')

heat.head(3)
```

```python
## Remover colunas com valores NaN e a coluna "country"
heat_cleaned = heat.drop('country', axis=1) #.dropna(axis=1)

## Renomear a coluna "age_group" para "faixas_etarias"
heat_cleaned = heat_cleaned.rename(columns={'age_group': 'faixas etárias'})

## Remover linhas com "Não Detectado" ou "Not tested"
heat_cleaned = heat_cleaned.loc[~heat_cleaned['DENV_test_result'].str.contains('Não Detectado|Not tested')]

## Substituir as faixas etárias no DataFrame
heat_cleaned['faixas etárias'] = heat_cleaned['faixas etárias'].replace({'0-4': '00-04', '4-9': '05-09', '9-19': '10-19', '19-29': '20-29', '29-39': '30-39', '39-49': '40-49', '49-59': '50-59', '59-69': '60-69', '69-79': '70-79'})

heat_cleaned.head(3)
```

```python
## Derreter (melt) as colunas de data em uma coluna
heat_melted = pd.melt(heat_cleaned, id_vars=['DENV_test_result', 'faixas etárias'], var_name='semana epidemiológica', value_name='percentual')
heat_melted.head(3)
```

```python
## Transformar os valores da coluna "n" em percentuais
heat_melted['percentual'] = heat_melted['percentual'].apply(lambda x: f'{round(x * 100, 2)}%')
#heat_melted = heat_melted.sort_values('faixas etárias')
heat_melted.head(3)
```

```python
## Filtrar os valores positivos e negativos
heat_melted_positivos = heat_melted[heat_melted['DENV_test_result'] == 'Pos'].reset_index(drop=True)
# heat_melted_negativos = heat_melted[heat_melted['DENV_test_result'] == 'Negative'].reset_index(drop=True)
heat_melted_positivos.head(3)
```

```python
## Escrever o DataFrame final positivo em um arquivo TSV
#heat_melted_positivos.to_csv('figures/_flourish/heatmap_DENVdemog.tsv', sep='\t', index=False)

## Escrever o DataFrame final positivo em um arquivo excel -
heat_melted_positivos.to_excel('heatmap_DENVdemog.xlsx', index=False)
```

### Estados

```python
## Carregar tabela
heat_ufs = pd.read_csv("figures/heatmap/combined_matrix_state_posrate_direct_weeks.tsv", sep='\t')
heat_ufs.head(3)
```

```python
## Correções pontuais
# Converter "RIO DE JANEIRO" para "Rio de Janeiro" em letras minúsculas
heat_ufs['state'] = heat_ufs['state'].replace({'rio de janeiro': 'Rio de Janeiro'})
heat_ufs['state_code'] = heat_ufs['state_code'].replace({'0': 'RJ'})
```

```python
melted_ufs = heat_ufs.melt(id_vars=['state', 'pathogen', 'test_result', 'state_code', 'country'],
                    var_name='semana epidemiológica',
                    value_name='percentual')
melted_ufs.head()
```

```python
## Transformar os valores da coluna "n" em percentuais
melted_ufs['percentual'] = melted_ufs['percentual'].apply(lambda x: f'{round(x * 100, 2)}')
#heat_melted = heat_melted.sort_values('faixas etárias')
melted_ufs.head(3)
```

```python
## Filtrar as linhas
heatmap_denv = melted_ufs[melted_ufs['pathogen'] == 'DENV']
```

```python
## Remover e renomear colunas
columns_to_remove = ['state', 'test_result', 'country', 'pathogen']
heatmap_uf_denv = heatmap_denv.drop(columns=columns_to_remove)

# Renomear coluna
heatmap_uf_denv = heatmap_uf_denv.rename(columns={'state_code': 'UF'})
heatmap_uf_denv.head(10)
```

```python
print(heatmap_uf_denv.loc[heatmap_uf_denv['UF'] == 'RJ'])
```

```python
## convertendo 'nan' para  NaN
heatmap_uf_denv.replace('nan', np.nan, inplace=True)

## Filtrar estados com mais de 80% dos dados preenchidos diferentes de "nan" na coluna "percentual" por estado
porcentagem = heatmap_uf_denv.groupby('UF')['percentual'].apply(lambda x: (x.notna().sum() / len(x)) * 100)

## Selecionar os estados com mais de 80% das linhas com dados diferentes de "nan"
estados_selecionados = porcentagem[porcentagem > 30].index

## Criar o novo DataFrame contendo apenas os estados selecionados
heatmap_ufs = heatmap_uf_denv[heatmap_uf_denv['UF'].isin(estados_selecionados)].dropna()

print(heatmap_ufs)
```

```python
print(heatmap_ufs.loc[heatmap_ufs['UF'] == 'RJ'])
```

```python
## ordenar estados - paliativo enquanto não ajusta para Snakemake
regioes = {
    'Norte': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS']
}

## Mapear as informações de região no DataFrame
heatmap_ufs['Regiao'] = heatmap_ufs['UF'].map({estado: regiao for regiao, estados in regioes.items() for estado in estados})

## Ordenar o DataFrame por região e em ordem alfabética
heatmap_ufs = heatmap_ufs.sort_values(['Regiao', 'UF'])

print(heatmap_ufs)
```

```python
## listar estados
print(heatmap_ufs['UF'].unique())
```

```python
## salvar excel
heatmap_ufs.to_excel('heatmap_states.xlsx', index=False)
```
